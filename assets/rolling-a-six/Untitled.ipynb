```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
p1, p2, p3, p4 = 1/2, 1/6, 1/20, 1/100
table = f""
table +=  "|Try $$t$$|Coin Flip|6-sided Dice|20-sided Dice|1 in 100|\n"
table += f"|---------|----------------------|-------------|--------|\n"
for t in range(1, 20+1):
    pr1 = (1-p1)**(t-1) * p1
    pr2 = (1-p2)**(t-1) * p2
    pr3 = (1-p3)**(t-1) * p3
    pr4 = (1-p4)**(t-1) * p4
    table += f"|$$T = {t}$$|$${pr1 * 100 : .1f}\%$$|$${pr2 * 100 : .1f} \%$$|$${pr3 * 100 : .1f} \%$$|$${pr4 * 100 : .1f} \%$$|\n"
print(table)
```

```python
p1, p2, p3, p4 = 1/2, 1/6, 1/20, 1/100
table = f""
table +=  "|Try $$t$$|Coin Flip|6-sided Dice|20-sided Dice|1 in 100|\n"
table += f"|---------|----------------------|-------------|--------|\n"
cpr1, cpr2, cpr3, cpr4 = 0, 0, 0, 0
for t in range(1, 30+1):
    pr1 = (1-p1)**(t-1) * p1
    pr2 = (1-p2)**(t-1) * p2
    pr3 = (1-p3)**(t-1) * p3
    pr4 = (1-p4)**(t-1) * p4
    cpr1 += pr1
    cpr2 += pr2
    cpr3 += pr3
    cpr4 += pr4
    table += f"|$$T \le {t}$$|$${cpr1 * 100 : .1f}\%$$|$${cpr2 * 100 : .1f} \%$$|$${cpr3 * 100 : .1f} \%$$|$${cpr4 * 100 : .1f} \%$$|\n"
print(table)
```

```python
t = np.arange(1, 105+1)
fig, ax = plt.subplots(figsize=(20, 10))
for p in [1/2, 1/6, 1/20, 1/100]:
    pr = (1-p)**(t-1) * p
    ax.plot(t, pr, '.', label=f'p = {p*100:.0f} %', markersize=14)
ax.set_yscale('log')
ax.legend(prop={'size': 20})
ax.set_title('Probabilities')
plt.tight_layout()
fig.savefig('pr.png', transparent=True)
```

```python
t = np.arange(1, 105+1)
fig, ax = plt.subplots(figsize=(20, 10))
for p in [1/2, 1/6, 1/20, 1/100]:
    pr = (1-p)**(t-1) * p
    cpr = np.cumsum(pr)
    ax.plot(t, cpr, '.', label=f'p = {p*100:.0f} %', markersize=14)
# ax.set_yscale('log')
ax.legend(prop={'size': 20})
ax.set_title('Cummulative Probabilities')
plt.tight_layout()
fig.savefig('cpr.png', transparent=True)
```

```python
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

t = np.arange(1, 105+1)
fig, ax = plt.subplots(figsize=(20, 10))
for i, p in enumerate([1/2, 1/6, 1/20, 1/100]):
    et = t * (1-p)**(t-1) * p
    ax.plot(t, et, '.', label=f'p = {p*100:.0f} %', markersize=14)
    ax.vlines(x=1/p, ymin=0, ymax=et[int(1/p)-1], color=colors[i])
# ax.set_yscale('log')
ax.legend(prop={'size': 20})
ax.set_title('Probabilities')
plt.tight_layout()
fig.savefig('weighted_terms.png', transparent=True)
```

```python
t = np.arange(1, 3000)
table = ""
table += "|Event|Probability|$$1$$ to $$\mathbb{E}$$|$$\mathbb{E}+1$$ to $$\infty$$|$$\mathbb{E}$$|\n"
table += "|-----|-----------|-------------------------|----------------------------|--------------|\n"
events = ["Coin flip", "6-sided dice", "20-sided dice", "1 in 100"]
for i, p in enumerate([1/2, 1/6, 1/20, 1/100]):
    et = t * (1-p)**(t-1) * p
    e = int(1//p) - 1 
    table += f"|{events[i]} |$${p * 100 : .0f}\%$$|$${np.sum(et[:e+1]) : .1f}$$|$${np.sum(et[e+1:]) : .1f}$$|$${np.round(np.sum(et), 0) : .0f}$$|\n"
print(table)
```

```python
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

t = np.arange(1, 100+1)
fig, ax = plt.subplots(figsize=(20, 10))
for i, p in enumerate([1/2, 1/6, 1/20]):
    et = t * (1-p)**(t-1) * p
    cet = np.cumsum(et)
    ax.plot(t, cet, '.', label=f'p = {p*100:.0f} %', markersize=14)
    ax.vlines(x=1/p, ymin=0, ymax=cet[int(1/p)-1], color=colors[i])
    ax.hlines(y=(1//p)/2, xmin=0, xmax=100, color=colors[i])

# ax.set_yscale('log')
ax.legend(prop={'size': 20})
ax.set_title('Probabilities')
fig.savefig('c_weighted_terms.png', transparent=True)
```
